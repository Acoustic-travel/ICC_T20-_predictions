```python
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = [os.path.join('t20s', file) for file in os.listdir('t20s')]
```

```python
dfs = []
```

```python
for counter, file in enumerate(tqdm(filenames, desc="Processing files")):
    with open(file, 'r') as f:
        try:
           
            data = safe_load(f)
            df = pd.json_normalize(data)
            
            df['match_id'] = counter + 1
            
            dfs.append(df)
        except Exception as e:
            print(f"Error processing file {file}: {e}")

final_df = pd.concat(dfs, ignore_index=True)
print(final_df)
```
